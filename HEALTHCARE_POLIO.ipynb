```python
#HEALTH CARE-POLIO
#To solve individual applications-visual Analytics for Healthcare

#he reported figures include both wild- and vaccine-derived type polio infections that occurred indigenously and as imported cases.

#Variable description:

#Year: from 1910 to 2019
#polio_ cases: Cases reported
#polio_ deaths: Deaths reported (OWID based on US Public Health Service (1910-1951) and US Center for Disease Control (1960-2010))
#Source: Our World In Data based on US Public Health Service; US Center for Disease Control; and WHO
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv('/content/usa-polio-cases.csv')
df.head()
```

```python
#summary
df.info()
```

```python
#THE MEAN, VARIANCE AND STANDARD DEVIATION FOR 	NUMBER OF POLIO CASE
mean = df['polio_cases'].mean()
var = df['polio_cases'].var()
std = df['polio_cases'].std()
print (f"Mean {mean}\nVariance {var}\nStandard Deviation = {std}")
```

```python
#THE MEAN, VARIANCE AND STANDARD DEVIATION FOR DEATH CASES IN POLIO DISEASE
mean = df['polio_deaths'].mean()
var = df['polio_deaths'].var()
std = df['polio_deaths'].std()
print (f"Mean {mean}\nVariance {var}\nStandard Deviation = {std}")
```

```python
plt.figure(figsize=(10,2))
plt.boxplot(df['Year'], vert=False, showmeans=True)
plt.grid(color='red', linestyle='dotted')
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.boxplot(df['polio_cases'], vert=False, showmeans=True)
plt.grid(color='purple', linestyle='dotted')
plt.show()
```

```python
#DISTRIBUTION OF POLIO ATTACKED PATIENT
df['polio_cases'].hist (bins=15)
plt.suptitle('Each year polio attacked death patient')
plt.xlabel('Year')
plt.ylabel('polio_deaths')
plt.show()
```

```python
plt.scatter(df['Year'],df['polio_deaths'])
plt.show()
```

```python
sns.histplot(df['polio_cases'], kde=True)
plt.title('Histogram with KDE')
plt.show()
```

```python
fig = px.scatter(df, x='Year', y='polio_cases', title='Scatter Plot')
fig.show()
```

```python
from bokeh.plotting import figure, show
p = figure(title='Bokeh Example', x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(df['polio_deaths'], df['polio_cases'], size=10)
show(p)
```
